# 함수와 메서드 차이 알아보기

- 내장함수 : 프로그램안에 미리 정의되어 있는 함수 
  sum,max,print,min
- 패키지 함수 : 별도로 설치, 불러와야하는 함수 
  import pandas as pd
  import matplotlib.pylot as plt
- 메서드 : 변수가 지니고 있는 함수 이름과 괄호를 같이 사용
  df.head().info().tail().shape().describe()
  메서드 = 기술, 어트리뷰트 = 능력치

```python
df_raw = pd.DataFrame({'Var1' : [1,2,1], 
                      'Var2' : [2,3,2]})
df_raw
```

```python
df_raw.shape
```

```python
# 데이터프레임 복사본 만들기 .copy()
df_new = df_raw.copy()
df_new
```

```python
# 변수명 바꾸는 메서드 .rename()
df_new = df_new.rename(columns = {'Var2' : 'V2'})
```

```python
df_new
```

```python
df_new = df_new.rename(columns = {'Var1' : 'V1'})
df_new
```

```python
print(df_raw)
print(df_new)
```
