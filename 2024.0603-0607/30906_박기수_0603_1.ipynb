```python
# 꺾은 선 그래프 그리기
# 라이브러리 불러오기 및 별명 지정하기

import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')

# x,y축 데이터 지정
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

# 그래프의 해상도 설정하기 dpi figure
plt.figure(dpi=100)

# 제목 title
plt.title('월별 판매 실적')

# 꺾은선형 그래프 그리기.plot.color
# 그래프 범례 .label
plt.plot(month,sales,color='yellow',label='단위:개수')

# 범례를 보이게 설정 위치.lower right
plt.legend(loc = 'lower right')
# x축, y축 xlabel, ylabel
plt.xlabel("월")
plt.ylabel('매출')

# 그래프 보이기
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')

# 데이터 지정하기
data = [1,2,3,4,5,6,7,8,9]
mon = [92,63,8,27,39,84,76,18,46]

plt.figure(dpi=100)

# 꺾은선형 색 red label 지출
plt.plot(data,mon,color='red',label='지출')

plt.legend(loc='upper left') # 위 / 왼


plt.show()
```
