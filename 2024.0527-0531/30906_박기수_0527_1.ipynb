```python
# 라이브러리 불러오기 및 별명 지정
import matplotlib.pyplot as plt

# 그래프용 데이터
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']
plt.figure(dpi=150)

# 한글 폰트
plt.rc('font',family="Malgun Gothic")
# 그래프 제목 넣기 plt.title
plt.title('준원 애미')
# 그래프 해상도 설정


# 그래프 그리기
plt.plot(month,sales,color='black') # 그래프 색 바꾸기 color="색이름"
plt.show()
```
